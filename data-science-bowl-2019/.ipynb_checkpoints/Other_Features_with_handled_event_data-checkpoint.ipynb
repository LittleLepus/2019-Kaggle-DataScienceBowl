```python
def from_list_to_dummies(df):
    for i in df.columns:
        dum=pd.get_dummies(i)
        
        df=pd.concat([df,dum],axis=1)
        df.drop(i,axis=1,inplace=True)
    return df
```

```python
def unique_list(list_d):
    u_l=[]
    for i in list_d:
        if i not in u_l:
            u_l.append(i)
    return u_l
```

```python
def add_freq_for_actions(df,id_list,id_col):
    o=0
    freq_df=pd.DataFrame(index=range(0,len(id_list)), columns=df.columns)
    for j in id_list:
        love=df[df[id_col]==j]
        #love.drop(id_col,axis=1,inplace=True)
        for i in df.columns:
            if i != id_col:
                freq_df[i][o]=sum(love[i])
        o=o+1
    return freq_df
```
