```python
def event_sorter(ent,cols):
    df=pd.DataFrame(index=range(0,len(ent)),columns=cols)
    c=df.columns
    num=0
    for i in ent:
        e=0
        #print(i)

        for k in i:
            cc=c[e]
            df[cc][num]=k
            e=e+1
            #print(k)
        num=num+1
    return df
```

```python
def unique_col_names(col_list,unique_ending):
    new_col_list=[]
    for col in col_list:
        n_col=col + "_" + unique_ending
        new_col_list.append(n_col)
    return new_col_list
```

```python
def sub_dict_parser(dict_list):
    for j in dict_list:
        if pd.notnull(j):
            k=[*j]
            break
    df=pd.DataFrame(index=range(0,len(dict_list)),columns=k)
    io=0
    for o in k:
        io=0
        for i in dict_list:
            if pd.notnull(i):
                df[o][io]=i.get(o)
                io=io+1
    return df
```

```python
def event_sorter(ent,cols):
    df=pd.DataFrame(index=range(0,len(ent)),columns=cols)
    c=df.columns
    num=0
    for i in ent:
        e=0
        #print(i)

        for k in i:
            cc=c[e]
            df[cc][num]=k
            e=e+1
            #print(k)
        num=num+1
    return df
```
