```python
def event_sorter(ent,cols):
    df=pd.DataFrame(index=range(0,len(ent)),columns=cols)
    c=df.columns
    num=0
    for i in ent:
        e=0
        #print(i)

        for k in i:
            cc=c[e]
            df[cc][num]=k
            e=e+1
            #print(k)
        num=num+1
    return df
```

```python
def unique_col_names(col_list,unique_ending):
    new_col_list=[]
    for col in col_list:
        n_col=col + "_" + unique_ending
        new_col_list.append(n_col)
    return new_col_list
```

```python
def sub_dict_parser(dict_list):
    for j in dict_list:
        if pd.notnull(j):
            k=[*j]
            break
    df=pd.DataFrame(index=range(0,len(dict_list)),columns=k)
    io=0
    for o in k:
        io=0
        for i in dict_list:
            if pd.notnull(i):
                df[o][io]=i.get(o)
                io=io+1
    return df
```

```python
def handle_strings(df):
    dum_df=df
    for i in final_game_df_all.columns:
        n=df[i]
        
        if (type(n.values[0]).__name__ =='str')|(type(n.values[0]).__name__ =='bool'):
            dum=pd.get_dummies(n)
            dum_df=pd.concat([dum_df,dum],axis=1)
            dum_df.drop(i,axis=1,inplace=True)
    return dum_df
```

```python
def unique_list(list_d):
    u_l=[]
    for i in list_d:
        if i not in u_l:
            u_l.append(i)
    return u_l
```

```python
def change_list_val_str(df,portion):
    emp_list=[]
    for i in df[portion]:
        emp_list.append(str(i))
    df[portion]=emp_list
    return df
```
