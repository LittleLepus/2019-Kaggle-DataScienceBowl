```python
import pickle as pkl
infile1 = open('freq_data.pkl','rb') 
freq_df=pkl.load(infile1)
infile1.close()
```

```python
infile2 = open('ultimate_df_pickle.pkl','rb')  
# load the object from the file into var b
event_data_df = pkl.load(infile2)
infile2.close()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler



X=complete_df.drop('target_76',axis=1)
X.replace('unknown',0,inplace=True)
y=complete_df['target_76']
y=y.astype('int')

inst_mm=MinMaxScaler()
fit_mm=inst_mm.fit_transform(X)

X_train,X_test,y_train,y_test=train_test_split(fit_mm,y,test_size=.25)
```

```python
fit_mm
```

```python
from xgboost import XGBClassifier
inst_xgb=XGBClassifier(
#learning_rate =0.1,
# n_estimators=1000,
# max_depth=5,
# min_child_weight=1,
# gamma=0,
# subsample=0.8,
# colsample_bytree=0.8,
# objective= 'binary:logistic',
# nthread=4,
 #scale_pos_weight=1,
 #seed=27
)
fit_xgb=inst_xgb.fit(X_train,y_train)
y_pred_xgb=fit_xgb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred_xgb,y_test)
cohen_kappa_score(y_pred_xgb,y_test)
#accuracy_score(y_pred_xgb,y_test)
```

```python
def thresh_for_regress(thresh_1,thresh_2,thresh_3,input_l):
    pred_trans=[]
    for i in input_l:
        if i<thresh_1:
            pred_trans.append(0)
        elif (i>=thresh_1) & (i<=thresh_2):
            pred_trans.append(1)
        elif (i>=thresh_2) & (i<=thresh_3):
            pred_trans.append(2)
        else:
            pred_trans.append(3)
    return pred_trans
```

```python
import lightgbm as lgb
param = {'num_leaves':150, 'objective':'binary','max_depth':7,'learning_rate':.05,'max_bin':200}
param['metric'] = ['auc', 'binary_logloss']
num_round=50
lgb_test=lgb.Dataset(X_test)
lgb_train=lgb.Dataset(X_train,label=y_train)
model=lgb.train(param,lgb_train,50)
y_pred_lgb=model.predict(X_test)
y_trans_pred_lgb=thresh_for_regress(.25,.5,.75,y_pred_lgb)
cohen_kappa_score(y_trans_pred_lgb,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
inst_rf=RandomForestClassifier()
fit_rf=inst_rf.fit(X_train,y_train)
y_pred_rf=fit_rf.predict(X_test)
cohen_kappa_score(y_pred_rf,y_test)
```
